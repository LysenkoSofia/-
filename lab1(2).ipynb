$x'=ax-bx^2$

Модель Мальтиуса с зависящим от населения коэффициентом размножения

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
def dxdt(x,t):
    a=0.6
    b=0.1
    return a*x-b*x*x
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=ax-bx^2$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=x-x^2-c$

Модель рыболовства с интенсивностью _с_, где х- количество рыб в океане

```python
def dxdt(x,t):
    c=0.8
    return x-x*x-c
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=x-x^2-c$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=x-g(x)x-c$

Модель рыболовства, где критическое значение _с_ является максимум функции _g(x)x_

```python
def dxdt(x,t):
    a=0.4
    b=1
    c=0.3
    return x-(a-b*x)*x-c
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=x-g(x)x-c$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=x-x^2-kx$

Модель изменения популяции рыб при дифференциальной квоте вылова

```python
def dxdt(x,t):
    k=8
    return x-x**2-k*x
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=x-x^2-kx$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=-kx$

Модель Мальтиуса(для роста населения Земли)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
def dxdt(x,t):
    k=0.42
    return -k*x
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=-kx$")
plt.grid(True)
plt.legend()
plt.show()
```
