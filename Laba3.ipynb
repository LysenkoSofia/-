$x'=-kx$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
def dxdt(x,t):
    k=0.42
    return -k*x
t=np.linspace(0,10,100)
solve=odeint(dxdt,y0=1,t=t)
solve=solve.flatten()
```

```python
%matplotlib inline
plt.plot(t,solve,label=r"$x'=-kx$")
plt.grid(True)
plt.legend()
plt.show()
```

```python
y0=1
h=0.1
t_el=np.arange(0,10,h)
solve2=[y0]
for t in t_el:
    yk=solve2[-1]+h*dxdt(t,solve2[-1])
    solve2.append(yk)
solve2=solve2[:-1]
```

```python
%matplotlib inline
plt.plot(t_el,solve2,label=r"$эйлер$")
plt.grid(True)
plt.legend()
plt.show()
```
